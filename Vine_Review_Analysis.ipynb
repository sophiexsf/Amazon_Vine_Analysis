```python
import pandas as pd
```

```python
df = pd.read_csv("vine_table.csv")
```

```python
df.head()
```

```python
filtered_df = df.loc[df.total_votes >= 20]
filtered_df.head()
```

```python
helpful_df = filtered_df.loc[filtered_df.helpful_votes / filtered_df.total_votes >= 0.5]
helpful_df.head(20)
```

```python
paid_reviews_df = helpful_df.loc[helpful_df.vine == 'Y']
unpaid_reviews_df = helpful_df.loc[helpful_df.vine == 'N']
```

```python
paid_total = len(paid_reviews_df)
unpaid_total = len(unpaid_reviews_df)

paid_5star = len(paid_reviews_df[paid_reviews_df.star_rating == 5.0])
unpaid_5star = len(unpaid_reviews_df[unpaid_reviews_df.star_rating == 5.0])
```

```python
result = pd.DataFrame({'vine': ['Y', 'N'], 'total': [paid_total, unpaid_total], '5-star': [paid_5star, unpaid_5star]})
result['5-star percentage'] = result['5-star'] * 100.0 / result['total']
result['5-star percentage'] = result['5-star percentage'].map("{:,.2f}%".format)
result
```
